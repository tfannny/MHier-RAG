系统测评总结流程(以MMLongBench-Doc数据集为例):
1. MMLongBench PDF -> 放入 pdf_reports 文件夹。
2. MMLongBench JSON -> 转换为 questions.json -> 运行 Pipeline -> 得到 answers.json。
3. answers.json + MMLongBench 真值 -> LLM 打分 -> 得到 final_scored_results.json。
4. final_scored_results.json -> 运行 test_eval_acc_f.py -> 得到最终准确率报告。
5. 
特别注意： 你的评测脚本 test_eval_acc_f.py 中有这样一行代码： sample["evidence_pages"] = eval(sample["evidence_pages"]) 这说明它期望 JSON 中的 evidence_pages 字段是一个字符串形式的列表 (例如String: "[1, 3]" )，而不是直接的列表对象。如果在之前的步骤中它已经是列表了，这里 eval 可能会报错，需要根据实际情况去掉 eval 或者调整数据格式。
